```python
import pandas as pd
import seaborn as sns
```

```python
avocado = pd.read_csv('avocado.csv')
avocado.head()
```

```python
avocado.info()
```

```python
avocado['Date'] = pd.to_datetime(avocado['Date'])
avocado.info()
```

```python
avocado_new = avocado.query('year == 2015 and type == "conventional" and region == "Albany"')
avocado_new = avocado[['Date','Total Bags','Small Bags']]
avocado_new.head()
```

```python
avocado_melt = avocado.melt(id_vars=['Date'],
                            value_vars=['Total Bags', 'Small Bags'],
                            var_name='Bags')
avocado_melt.rename(columns={'value':'Count'}, inplace=True)
avocado_melt
```

```python
sns.lineplot(data=avocado_melt,x='Date',y='Count',hue='Bags')
```

```python
avocado["Sales Volume"] = pd.qcut(avocado["Total Volume"],q=4,labels=['Poor','Modest','Good','Excellent'])
avocado.head()
```

```python
sns.countplot(data=avocado, x='year', hue='Sales Volume')
```

```python
avocado_melted = avocado.melt(id_vars=['year'],
                              value_vars=['Small Bags','Large Bags','XLarge Bags'],
                              var_name='Bags',
                              value_name='Count')
                                
avocado_melted
```

```python
sns.relplot(data=avocado_melted, kind='line',x='year',y='Count',hue='Bags')
```
